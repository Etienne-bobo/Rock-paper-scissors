```python
import pyttsx3
import random

elements_list = ['Rock', 'Paper', 'Scissors']
textToSay = ''
user_score = 0
computer_score = 0
user_choice = ''
computer_choice = ''
gender = ''
name = ''

def start_announcement():
    global gender
    global name
    print("Hello, Wanna play Rock Paper and Scissor Game. Let's Start this. Good Luck!!")
    name = input("Enter your name: ")
    gender = input("Enter your gender: ")

def computer_action():
    choise = random.choices(elements_list)
    for x in range(len(choise)):
        return choise[x]
    
def text_to_speech_converter(textToSay):
    global gender
    engine = pyttsx3.init()
    engine.setProperty("rate", 180)
    voices = engine.getProperty('voices')
    if(gender == 'male'):
        engine.setProperty('voice', voices[0].id)

    elif(gender == 'female'):
        engine.setProperty('voice', voices[1].id)
    engine.say(textToSay)

    engine.runAndWait()
    
def print_announce_score():
    global user_score
    global computer_score
    if(computer_choice == 'Rock' ):
        if(user_choice == 'Paper'):
            user_score +=1
            print('You win this round! Paper cover Rock')
            text_to_speech_converter('You win this round Paper cover Rock')
            print('score after this round is player:', user_score, 'and computer: ', computer_score )
            text_to_speech_converter('score after this round is player'+ str(user_score) + 'and computer ' + str(computer_score) )
        elif(user_choice == 'Rock'):
            print('Tie!!')
            text_to_speech_converter('Tie')
            print('score after this round is player:', user_score, 'and computer: ', computer_score )
            text_to_speech_converter('score after this round is player'+ str(user_score) + 'and computer ' + str(computer_score) )

        elif(user_choice == 'Scissors'):
            computer_score +=1
            print('You lose this round! Rock smashes Scissors')
            text_to_speech_converter('You lose this round  Rock smashes Scissors')
            print('score after this round is player:', user_score, 'and computer: ', computer_score )
            text_to_speech_converter('score after this round is player'+ str(user_score) + 'and computer ' + str(computer_score) )
    
    elif(computer_choice == 'Paper'):
        if(user_choice == 'Rock'): 
            computer_score +=1
            print('You lose this round! Paper cover Rock')
            text_to_speech_converter('You lose this round  Paper cover Rock')
            print('score after this round is player:', user_score, 'and computer: ', computer_score )
            text_to_speech_converter('score after this round is player'+ str(user_score) + 'and computer ' + str(computer_score) )
        elif(user_choice == 'Paper'):
            print('Tie!!')
            text_to_speech_converter('Tie')
            print('score after this round is player:', user_score, 'and computer: ', computer_score )
            text_to_speech_converter('score after this round is player'+ str(user_score) + 'and computer ' + str(computer_score) )

        elif(user_choice == 'Scissors'):
            user_score +=1
            print('You win this round! Scissors cut paper')
            text_to_speech_converter('You win this round Scissors cut paper')
            print('score after this round is player:', user_score, 'and computer: ', computer_score )
            text_to_speech_converter('score after this round is player'+ str(user_score) + 'and computer ' + str(computer_score) )
    
    elif(computer_choice == 'Scissors'):
        if(user_choice == 'Rock'):
            user_score +=1
            print('You win this round! Rock smashes Scissors')
            text_to_speech_converter('You win this round  Rock smashes Scissors')
            print('score after this round is player:', user_score, 'and computer: ', computer_score )
            text_to_speech_converter('score after this round is player'+ str(user_score) + 'and computer ' + str(computer_score) )

        elif(user_choice == 'Scissors'):
            print('Tie!!')
            text_to_speech_converter('Tie')
            print('score after this round is player:', user_score, 'and computer: ', computer_score )
            text_to_speech_converter('score after this round is player'+ str(user_score) + 'and computer ' + str(computer_score) )

        elif(user_choice == 'Paper'):
            computer_score +=1
            print('You lose this round! Scissors cut paper')
            text_to_speech_converter('You lose this round  Scissors cut paper')
            print('score after this round is player:', user_score, 'and computer: ', computer_score )
            text_to_speech_converter('score after this round is player'+ str(user_score) + 'and computer ' + str(computer_score) )


start_announcement()
text_to_speech_converter('Welcome')
text_to_speech_converter("Let's start this game, Good luck")

u = 0
c = 0
exit = 'y'
while(exit == 'y' or exit == 'Y'):
    
    text_to_speech_converter('Select your action 0 for Rock  1 for Paper and 2 for Scissors')
    while(c < 3 and u < 3):
        try:
            user_input = int(input('Select your action 0. Rock, 1. Paper, 2. Scissors: '))
        except ValueError:
            print("Oops!")
            text_to_speech_converter('Oops!')
            
        if(user_input == 0):
            user_choice = 'Rock'
            print('Player choosed:', user_choice)
            text_to_speech_converter('Player choosed'+ user_choice)
            computer_choice = computer_action()
            print('Computer choosed:', computer_choice )
            text_to_speech_converter('Computer choosed'+ computer_choice)
            print_announce_score()
            c = computer_score
            u = user_score
        elif(user_input == 1):
            user_choice = 'Paper'
            print('Player choosed:', user_choice)
            text_to_speech_converter('Player choosed'+ user_choice)
            computer_choice = computer_action()
            print('Computer choosed:', computer_choice )
            text_to_speech_converter('Computer choosed'+ computer_choice)
            print_announce_score()
            c = computer_score
            u = user_score
        elif(user_input == 2):
            user_choice = 'Scissors'
            print('Player choosed:', user_choice)
            text_to_speech_converter('Player choosed'+ user_choice)
            computer_choice = computer_action()
            print('Computer choosed:', computer_choice )
            text_to_speech_converter('Computer choosed'+ computer_choice)
            print_announce_score()
            c = computer_score
            u = user_score
        else:
            print("That's not a valid play. Select correct action")
            text_to_speech_converter("That's not a valid play. Select correct action")
    else:
        if(computer_score > user_score):
            print('You lose this game, better luck next time....')
            text_to_speech_converter('You lose this game, better luck next time')
        else:
            print('Congratulation you win this game.....')
            text_to_speech_converter('Congratulation you win this game')
        u = c = computer_score = user_score = 0
        

    text_to_speech_converter('would you like to continue? if yes press y otherwise press any key to stop the game ')
    exit = input("would you like to continue? if yes press 'y' otherwise press any key to stop the game: ")
```
